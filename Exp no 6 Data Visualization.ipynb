```python
#importing the basic library
import numpy as np 
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

# Line Chart

```python
plt.plot(x,y)
```

```python
plt.title("Line chart")
```

```python
plt.xlabel("x axis")
```

```python
plt.ylabel("Y axis")
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")

plt.ylabel("Y axis")
plt.show()
```

```python
plt.bar(x,y,color='gray')
```

# Scatter Plot

```python
a=(1,2,5,6,7,8,6,3)
b=(12,13,14,15,16,17,18,19)
plt.scatter(a,b)
plt.title("scatter chart")
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(a,b,color='green')
```

```python
labels=['11','27','20','18','22','67']
sizes=[30,25,25,20,34,20]
plt.pie(sizes,labels=labels)
plt.title("Roll no")
plt.show()
```
